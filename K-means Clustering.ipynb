K-means:
El K-means es un algoritmo de clustering, que es una técnica de aprendizaje no supervisado utilizada para agrupar datos similares en conjuntos. El objetivo del K-means es dividir un conjunto de datos en grupos, llamados "clusters", de manera que los puntos dentro de un mismo cluster sean más similares entre sí que con los puntos de otros clusters. Este algoritmo busca minimizar la varianza intra-cluster y maximizar la varianza inter-cluster.

Qué es la técnica de aprendizaje no supervisado:es un enfoque en el campo del aprendizaje automático donde el algoritmo se entrena en datos que no están etiquetados, es decir, no se proporciona una respuesta conocida o salida deseada.

Clustering (Agrupamiento): Enfoque que agrupa los datos en conjuntos, llamados "clusters", de manera que los elementos dentro de un mismo cluster sean más similares entre sí que con los elementos de otros clusters.
(agrupar datos en categorías basadas en similitudes.)


Etiquetas (labels):se refieren a las asignaciones de clusters para cada punto de datos en el conjunto. Después de aplicar el algoritmo K-means a un conjunto de datos, cada punto se asigna a un cluster específico, y estas asignaciones se almacenan en el array labels. Cada valor en labels indica a qué cluster pertenece el punto correspondiente en el conjunto de datos. Por ejemplo, si labels[0] es igual a 1, significa que el primer punto del conjunto de datos pertenece al cluster 1.

Centroides:Los "centroides" son puntos representativos que sirven como el centro geométrico de un cluster. Después de aplicar K-means, el algoritmo ajusta los centroides para minimizar la distancia entre los puntos dentro del mismo cluster. Los centroides se almacenan en el array centroids. Cada fila en centroids contiene las coordenadas de un centroide, y el número de filas es igual al número de clusters (k).


EJEMPLO
Agrupar datos bidimensionales en tres clusters y visualiza los resultados para entender cómo se distribuyen los datos y dónde se encuentran los centroides de cada cluster.


```python
#1. Importar libreria np
import numpy as np
# 2.Importar la clase KMeans del módulo sklearn.cluster:
# from sklearn.cluster import KMeans: Aquí, estamos importando la clase KMeans desde el módulo sklearn.cluster de la biblioteca scikit-learn. Esta biblioteca proporciona herramientas eficientes para el aprendizaje de máquinas y la minería de datos.
from sklearn.cluster import KMeans
#3 Importando la biblioteca matplotlib.pyplot para visualizar los resultados.
import matplotlib.pyplot as plt

# Generar datos de ejemplo
#np.random.seed(42): Establece la semilla del generador de números aleatorios de numpy. Esto garantiza que los números aleatorios 
#generados sean los mismos en cada ejecución.
np.random.seed(42)

#4. Crea una matriz X de dimensiones 100x2 con valores aleatorios entre 0 y 1. Estos serán los datos de ejemplo en un plano bidimensional.
X = np.random.rand(100, 2)

# Definir el número de clusters (K)
k = 3

# 5.Aplicar el algoritmo K-means
#Crea una instancia del objeto KMeans con el número de clusters especificado.
kmeans = KMeans(n_clusters=k)
# kmeans.fit(X): Aplica el algoritmo K-means a los datos X que hemos generado. Este paso ajusta el modelo a los datos y 
#calcula los centroides de los clusters.
kmeans.fit(X)

# 6.Obtener las etiquetas de los clusters y los centroides
# labels = kmeans.labels_: Obtiene las etiquetas asignadas a cada punto de datos, indicando a qué cluster pertenecen.
labels = kmeans.labels_
 #Obtiene las coordenadas de los centroides de cada cluster.
centroids = kmeans.cluster_centers_

# Visualizar los resultados
#Define colores para los clusters en el gráfico.
colors = ['r', 'g', 'b']
#Itera sobre cada cluster para visualizar los puntos asociados a ese cluster en el gráfico.
for i in range(k):
    #Realiza un gráfico de dispersión de los puntos de un cluster específico con un color determinado.
    plt.scatter(X[labels == i, 0], X[labels == i, 1], c=colors[i], label=f'Cluster {i + 1}')

    #Agrega los centroides al gráfico.
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, linewidths=3, color='black', label='Centroids')
plt.title('Resultado del K-means Clustering')
#Etiqueta el eje x.
plt.xlabel('Característica 1')
#Etiqueta el eje y
plt.ylabel('Característica 2')
#Muestra la leyenda en el gráfico.
plt.legend()
#Muestra el gráfico.
plt.show()
```

K-Means es un algoritmo no supervisado de Clustering. Se utiliza cuando tenemos un montón de datos sin etiquetar. El objetivo de este algoritmo es el de encontrar “K” grupos (clusters) entre los datos crudos. 

**¿Como funciona K-Means**

El algoritmo K-means opera iterativamente asignando cada punto a uno de los K grupos según la similitud de sus características. 

Cada grupo tiene un centroide, una especie de coordenada representativa. El algoritmo ajusta orgánicamente la posición de estos centroides en cada iteración hasta que converge. 

Al final, obtenemos los centroides y etiquetas para el conjunto de datos de entrenamiento. Estos centroides son puntos clave que caracterizan cada grupo, y las etiquetas asignan a cada punto a uno de estos grupos.

Analizando los centroides, podemos identificar las características únicas de cada grupo. En resumen, K-means agrupa datos de manera iterativa, definiendo centroides representativos y asignando etiquetas a cada punto basándose en la similitud de características.

**El Algoritmo K-means** Alejandro

El algoritmo utiliza una proceso iterativo en el que se van ajustando los grupos para producir el resultado final. Para ejecutar el algoritmo deberemos pasar como entrada el conjunto de datos y un valor de K. El conjunto de datos serán las características o features para cada punto. Las posiciones iniciales de los K centroids serán asignadas de manera aleatoria de cualquier punto del conjunto de datos de entrada. Luego se itera en dos pasos:

1- Paso de Asignación de datos

En este paso, cada “fila” de nuestro conjunto de datos se asigna al centroide más cercano basado en la distancia cuadrada Euclideana. Se utiliza la siguiente fórmula (donde dist() es la distancia Euclideana standard):


2-Paso de actualización de Centroid

En este paso los centroid de cada grupo son recalculados. Esto se hace tomando una media de todos los puntos asignados en el paso anterior.



El algoritmo itera entre estos pasos hasta cumplir un criterio de detención:
*  si no hay cambios en los puntos asignados a los grupos,
* o si la suma de las distancias se minimiza,
* o se alcanza un número máximo de iteraciones.

El algoritmo converge a un resultado que puede ser el óptimo local, por lo que será conveniente volver a ejecutar más de una vez con puntos iniciales aleatorios para confirmar si hay una salida mejor. Recuerden siempre seguir

Definimos la entrada

Concretamos la estructura de datos que utilizaremos para alimentar el algoritmo. Como se ve, sólo cargamos las columnas op, ex y ag en nuestra variable X.

```python

X = np.array(dataframe[["op","ex","ag"]])
y = np.array(dataframe['categoria'])
X.shape
```

Ahora veremos una gráfica en 3D con 9 colores representando las categorías.

```python
fig = plt.figure()
ax = Axes3D(fig)
colores=['blue','red','green','blue','cyan','yellow','orange','black','pink','brown','purple']
asignar=[]
for row in y:
    asignar.append(colores[row])
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
```

Obtener el valor K
Vamos a hallar el valor de K haciendo una gráfica e intentando hallar el “punto de codo” que comentábamos antes. Este es nuestro resultado:

```python
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

Ejecutamos K-Means
Ejecutamos el algoritmo para 5 clusters y obtenemos las etiquetas y los centroids.

```python

kmeans = KMeans(n_clusters=5).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)
```

Ahora veremos esto en una gráfica 3D con colores para los grupos y veremos si se diferencian: (las estrellas marcan el centro de cada cluster)

Aqui podemos ver que el Algoritmo de K-Means con K=5 ha agrupado a los 140 usuarios Twitter por su personalidad, teniendo en cuenta las 3 dimensiones que utilizamos: Openess, Extraversion y Agreeablenes. Pareciera que no hay necesariamente una relación en los grupos con sus actividades de Celebrity.



```python
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
colores=['red','green','blue','cyan','yellow']
asignar=[]
for row in labels:
    asignar.append(colores[row])
 
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores, s=1000)
```
